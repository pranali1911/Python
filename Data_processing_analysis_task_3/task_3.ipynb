### Problem Statement 2: Search Query Keyword Analysis
 Problem Statement 2: Search Query Keyword Analysis


```python
query = "Buy mobile phone buy phone online"
```

```python
#Convert the input to lowercase.
# (no need to import replace from os)
# from os import replace


query = query.lower()
print("Lowercase Query:", query)

#Ignore common punctuation.
query = query.replace(".", "").replace(",", "").replace("!", "").replace("?", "")
print("Punctuation Removed Query:", query)

#Count the frequency of each keyword.
keyword = query.split()
for word in keyword:
    frequency = keyword.count(word)
    print(f"Keyword: '{word}', Frequency: {frequency}")
    


#Display only keywords searched more than once.
keywords_count = []
for kw in keyword:
    count = keyword.count(kw)
    if count > 1 and kw not in keywords_count:
        keywords_count.append(kw)
print("Keywords Searched More Than Once:", keywords_count)
    
```

## Problem Statement 4: Email Domain Usage Analysis


```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
```

```python
# Count how many users belong to each email domain.
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1
print("Email Domain Counts:", domain_count)
```

```python
# Calculate the percentage usage of each domain.

total_emails = len(emails)
domain_percentage = {}
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"Domain: '{domain}', Percentage Usage: {percentage:.2f}%")
    
```

## Problem Statement 5: Sales Spike Detection


```python

sales = [1200, 1500, 900, 2200, 1400, 3000]
```

```python
#Calculate the average daily sales.
avg = sum(sales) / len(sales)
print("Average Daily Sales:", avg)
#Detect days where sales are more than 30% above average.

for sale in sales:
    if sale > avg * 1.3:
        print("Sales more than 30% above average:", sale)

#Display the day number and sale value.

for i in range(len(sales)):
    if sales[i] > avg * 1.3:
        print(f"Day {i+1}: Sale Value: {sales[i]}")
        
```

## Problem Statement 6: Duplicate User ID Detection


```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
```

```python
#Identify duplicate user IDs.

duplicates = []
 
for i in user_ids:
    if user_ids.count(i) > 1 and i not in duplicates:
        duplicates.append(i)
        print(i)
```

```python
#Display how many times each duplicate appears.

for i in duplicates:
    count = user_ids.count(i)
    print(f"User ID: '{i}', Count: {count}")    
```
